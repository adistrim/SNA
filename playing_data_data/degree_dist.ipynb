```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

with open("facebook_combined.txt", "r") as file:
    for line in file:
        source, target = map(int, line.strip().split())
        G.add_edge(source, target)

plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=False, node_color='skyblue', node_size=10, edge_color='gray', linewidths=0.5)
plt.title("Facebook Friendship Network")
plt.show()
```

```python
# Calculate degree distribution
degree_sequence = [d for n, d in G.degree()]

# Count the occurrences of each degree
degree_counts = nx.degree_histogram(G)

# Plotting degree distribution
plt.bar(range(len(degree_counts)), degree_counts, width=0.8, color='skyblue')

plt.title("Degree Distribution of the Facebook Friendship Network")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
# Total number of nodes in the graph
total_nodes = len(G.nodes())

# Calculate the probability of each degree
degree_probability = [count / total_nodes for count in degree_counts]

# Plotting degree distribution
plt.bar(range(len(degree_probability)), degree_probability, width=0.8, color='skyblue')

# Add title and labels
plt.title("Degree Probability Distribution of the Facebook Friendship Network")
plt.xlabel("Degree")
plt.ylabel("Probability")

# Show plot
plt.show()
```

```python
closeness_centrality = nx.closeness_centrality(G)

# Plot the closeness centrality
plt.figure(figsize=(10, 6))
plt.bar(closeness_centrality.keys(), closeness_centrality.values(), color='skyblue')
plt.xlabel('Node')
plt.ylabel('Closeness Centrality')
plt.title('Closeness Centrality of the Dataset')
plt.show()
```

```python
betweenness_centrality = nx.betweenness_centrality(G)

plt.figure(figsize=(10, 6))
plt.bar(betweenness_centrality.keys(), betweenness_centrality.values(), color='skyblue')
plt.xlabel('Node')
plt.ylabel('Betweenness Centrality')
plt.title('Betweenness Centrality of the Dataset')
plt.show()
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G)

plt.figure(figsize=(10, 6))
plt.bar(eigenvector_centrality.keys(), eigenvector_centrality.values(), color='skyblue')
plt.xlabel('Node')
plt.ylabel('Eigenvector Centrality')
plt.title('Eigenvector Centrality of the Dataset')
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)

plt.figure(figsize=(10, 6))
plt.bar(degree_centrality.keys(), degree_centrality.values(), color='skyblue')
plt.xlabel('Node')
plt.ylabel('Degree Centrality')
plt.title('Degree Centrality of the Dataset')
plt.show()
```
