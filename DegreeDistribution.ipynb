```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Load the Zachary Karate Club network
G = nx.karate_club_graph()

# Plot the graph
pos = nx.spring_layout(G)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=300, edge_color='k', linewidths=0.5)

# Add title
plt.title("Zachary Karate Club Network")

# Show plot
plt.show()
```

```python
# Load the Zachary Karate Club network
G = nx.karate_club_graph()

# Calculate degree distribution
degree_sequence = [d for n, d in G.degree()]

# Count the occurrences of each degree
degree_counts = nx.degree_histogram(G)

# Plotting degree distribution
plt.bar(range(len(degree_counts)), degree_counts, width=0.8, color='skyblue')

# Add title and labels
plt.title("Degree Distribution of the Zachary Karate Club Network")
plt.xlabel("Degree")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

```python
# Total number of nodes in the graph
total_nodes = len(G.nodes())

# Calculate the probability of each degree
degree_probability = [count / total_nodes for count in degree_counts]

# Plotting degree distribution
plt.bar(range(len(degree_probability)), degree_probability, width=0.8, color='skyblue')

# Add title and labels
plt.title("Degree Probability Distribution of the Zachary Karate Club Network")
plt.xlabel("Degree")
plt.ylabel("Probability")

# Show plot
plt.show()
```
